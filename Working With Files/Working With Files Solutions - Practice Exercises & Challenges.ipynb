### 4. Solution(s): Print the First n Lines of a File

```python
# 4.1. Expected Output: 
try:
    num_lines_to_print = int(input('Enter the number of lines to print: '))
    if num_lines_to_print < 0:
        raise ValueError("Negative number entered.")
except ValueError as e:
    print(f"Invalid input: {e}")


# 4.2. Expected Output: 
def print_first_n_lines(filename, n):
    try:
        with open(filename, 'r') as file:
            lines = file.readlines()
            if n > len(lines):
                print(f"Please enter a valid number of lines. This file has {len(lines)} lines.")
            else:
                for i in range(n):
                    print(lines[i].strip('\n'))
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")

print_first_n_lines('This Text File.txt', 3)



# 4.3. Expected Output: 
with open('This Text File.txt', 'r') as file:
    lines = [line for line in file if line.strip()]
    for i in range(min(num_lines_to_print, len(lines))):
        print(lines[i].strip('\n'))


# 4.4. Expected Output: 
try:
    num_lines_to_print = int(input('Enter the number of lines to print: '))
except ValueError:
    num_lines_to_print = 5

with open('This Text File.txt', 'r') as file:
    lines = file.readlines()
    for i in range(min(num_lines_to_print, len(lines))):
        print(lines[i].strip('\n'))


# 4.5. Expected Output: 
num_lines_to_print = int(input('Enter the number of lines to print: '))

with open('This Text File.txt', 'r') as file:
    for i, line in enumerate(file):
        if i >= num_lines_to_print:
            break
        print(line.strip('\n'))
```

### 5.  Solution(s): Copy the Content of a File to Another File

```python
# 5.1. Expected Ouptut: 
try:
    with open(the_file, 'r') as read_file, open('Copy to Me.txt', 'w') as write_file:
        for line in read_file:
            write_file.write(line)
except FileNotFoundError:
    print(f"Error: The file '{the_file}' does not exist.")

    
# 5.2. Expected Ouptut: 
def copy_file(source, destination):
    try:
        with open(source, 'r') as read_file, open(destination, 'w') as write_file:
            for line in read_file:
                write_file.write(line)
    except FileNotFoundError:
        print(f"Error: File '{source}' not found.")

copy_file('This Text File.txt', 'Copy to Me.txt')

    
# 5.3. Expected Ouptut: 
with open(the_file, 'r') as read_file, open('Copy to Me.txt', 'w') as write_file:
    for line in read_file:
        if line.strip():
            write_file.write(line)


# 5.4. Expected Ouptut: 
keyword = 'specific'
with open(the_file, 'r') as read_file, open('Copy to Me.txt', 'w') as write_file:
    for line in read_file:
        if keyword in line:
            write_file.write(line)
    
    
# 5.5. Expected Ouptut: 
chunk_size = 1024  # 1 KB
with open(the_file, 'r') as read_file, open('Copy to Me.txt', 'w') as write_file:
    while chunk := read_file.read(chunk_size):
        write_file.write(chunk)
```

### 6. Solution(s): Count Characters in a File

```python
# 6.1. Expected Ouptut: 
# Define the file to read
the_file = 'This Text File.txt'

# Open the file and count characters, including spaces
with open(the_file, 'r') as read_file:
    num_of_chars = 0
    for line in read_file:
        num_of_chars += len(line.strip('\n'))  # Spaces are included
print(num_of_chars)



# 6.2. Expected Ouptut: 
def count_chars(filename):
    try:
        with open(filename, 'r') as file:
            return sum(len(line.replace(' ', '').strip('\n')) for line in file)
    except FileNotFoundError:
        return "File not found."

print(count_chars('This Text File.txt'))


# 6.3. Expected Ouptut: 
import string

# Define the file to read
the_file = 'This Text File.txt'

# Open the file and count characters, excluding punctuation
with open(the_file, 'r') as read_file:
    num_of_chars = 0
    for line in read_file:
        cleaned_line = line.replace(' ', '').strip('\n')
        cleaned_line = cleaned_line.translate(str.maketrans('', '', string.punctuation))
        num_of_chars += len(cleaned_line)
print(num_of_chars)


# 6.4. Expected Ouptut: 
# Define the file to read
the_file = 'This Text File.txt'

# Open the file and count only alphabetic characters
with open(the_file, 'r') as read_file:
    num_of_chars = 0
    for line in read_file:
        num_of_chars += sum(char.isalpha() for char in line)  # Count alphabetic characters
print(num_of_chars)


# 6.5. Expected Ouptut: 
# Define the file to read
the_file = 'This Text File.txt'

try:
    with open(the_file, 'r') as read_file:
        num_of_chars = sum(len(line.replace(' ', '').strip('\n')) for line in read_file)
    print(num_of_chars)
except FileNotFoundError:
    print("The file does not exist.")
```

### 7. Solution(s): Find the Longest Word in a File

```python
# 7.1. Expected Output: 
longest_word = ''

with open(the_file, 'r') as read_file:
    for line in read_file:
        for word in line.split():
            if len(word.lower()) > len(longest_word.lower()):
                longest_word = word

print(longest_word)


# 7.2. Expected Output: 
def find_longest_word(filename):
    longest_word = ''
    try:
        with open(filename, 'r') as file:
            for line in file:
                for word in line.split():
                    if len(word) > len(longest_word):
                        longest_word = word
        return longest_word
    except FileNotFoundError:
        return "File not found."

print(find_longest_word('This Text File.txt'))


    
# 7.3. Expected Output: 
import string

longest_word = ''

with open(the_file, 'r') as read_file:
    for line in read_file:
        for word in line.split():
            # Remove punctuation from the word
            word = word.translate(str.maketrans('', '', string.punctuation))
            if len(word) > len(longest_word):
                longest_word = word

print(longest_word)


# 7.4. Expected Output: 
longest_words = []
max_length = 0

with open(the_file, 'r') as read_file:
    for line in read_file:
        for word in line.split():
            if len(word) > max_length:
                max_length = len(word)
                longest_words = [word]
            elif len(word) == max_length:
                longest_words.append(word)

print(longest_words)

    
# 7.5. Expected Output: 
try:
    with open(the_file, 'r') as read_file:
        longest_word = ''
        for line in read_file:
            for word in line.split():
                if len(word) > len(longest_word):
                    longest_word = word
        print(longest_word)
except FileNotFoundError:
    print("The file does not exist.")
```
