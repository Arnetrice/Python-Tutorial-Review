### 3. Solution(s): Remove Matching Element

```python
# 3.1. Expected Output: [1, 2, 3, 4]
the_list = [1, 2, 5, 3, 5, 4]
remove_this = 5
the_list = list(filter((remove_this).__ne__, the_list))
print(the_list)


# 3.2. Expected Output: ["1", 2, "2"]
the_list = [1, "1", 2, "2"]
remove_this = 1
the_list = [x for x in the_list if x != remove_this]
print(the_list)


# 3.3. Expected Output: ["banana", "cherry"]
the_list = ["apple", "banana", "apple", "cherry"]
remove_this = "apple"
the_list = list(filter((remove_this).__ne__, the_list))
print(the_list) 

# 3.4. Expected Output: ["b", "c"]
def remove_element(lst, elem):
    return list(filter((elem).__ne__, lst))

print(remove_element(["a", "b", "a", "c"], "a"))


# 3.5. Expected Output: ["a"]
the_list = [["a"], "a", ["a"]]
remove_this = ["a"]
the_list = [x for x in the_list if x != remove_this]
print(the_list)
```

### 6. Solution(s): Multiply All Elements in a List

```python
# 6.1. Expected Ouptut: [6, 12, 18, 24]
the_list = [2, 4, 6, 8]
the_factor = 3
for i, elem in enumerate(the_list):
    the_list[i] = elem * the_factor
print(the_list)


# 6.2. Expected Ouptut: [4, 8, 12]
def multiply_list(lst, factor):
    for i, elem in enumerate(lst):
        lst[i] = elem * factor
    return lst

print(multiply_list([1, 2, 3], 4))


# 6.3. Expected Ouptut: [0.5, 1.0, 1.5]
the_list = [1, 2, 3]
the_factor = 0.5
for i, elem in enumerate(the_list):
    the_list[i] = elem * the_factor
print(the_list)


# 6.4. Expected Ouptut: [[2, 4], [6, 8]]
the_list = [[1, 2], [3, 4]]
the_factor = 2
for sublist in the_list:
    for i, elem in enumerate(sublist):
        sublist[i] = elem * the_factor
print(the_list)


# 6.5. Expected Ouptut: [6, 8, 10, 12]
the_list = [3, 4, 5, 6]
the_factor = 2
the_list = [elem * the_factor for elem in the_list]
print(the_list)
```

### 7. Solution(s): Flatten a List that Contains Lists

```python
# 7.1. Expected Output: [1, 2, 3, 4, 5, 6, 7] for both
the_nested_list = [[1, 2, 3], 4, [5, 6], 7]

# Standard Approach
the_flatten_list = []
for i in the_nested_list:
    if not isinstance(i, list):
        the_flatten_list.append(i)
    else:
        for j in i:
            the_flatten_list.append(j)
print(the_flatten_list)
# Expected output: [1, 2, 3, 4, 5, 6, 7]

# Optimized Solution
the_flatten_list = [
    item for sublist in the_nested_list
    for item in (sublist if isinstance(sublist, list) else [sublist])
]
print(the_flatten_list)


# 7.2. Expected Output: ['a', 'b', 4, 5, 6]
def flatten_list(lst):
    return [
        item for sublist in lst
        for item in (sublist if isinstance(sublist, list) else [sublist])
    ]

print(flatten_list([['a', 'b'], 4, [5, 6]]))

    
# 7.3. Expected Output: [1, 2, 3, 4]
the_nested_list = [[1, 2, None], [3, None], 4]
the_flatten_list = [
    item for sublist in the_nested_list
    for item in (sublist if isinstance(sublist, list) else [sublist])
    if item is not None
]
print(the_flatten_list)


# 7.4. Expected Output: [1, 2, 3]
the_nested_list = [[], [1, 2], [], 3]
the_flatten_list = [
    item for sublist in the_nested_list
    if sublist != [] for item in (sublist if isinstance(sublist, list) else [sublist])
]
print(the_flatten_list)

    
# 7.5. Expected Output: [1, 2, 3]
def recursive_flatten(lst):
    flat_list = []
    for i in lst:
        if isinstance(i, list):
            flat_list.extend(recursive_flatten(i))
        else:
            flat_list.append(i)
    return flat_list

print(recursive_flatten([[1, [2, [3]]]]))
```

### 8. Solution(s): Find the nth Largest Value in a List

```python
# 8.1. Expected Output: 7
the_list = [4, 8, 2, 7, 9, 1]
nth_largest_value = 3
the_list.sort(reverse=True)
print(the_list[nth_largest_value - 1])


# 8.2. Expected Output: None
the_list = [3, 3, 3]
nth_largest_value = 2
if len(set(the_list)) >= nth_largest_value:
    sorted_list = sorted(set(the_list), reverse=True)
    print(sorted_list[nth_largest_value - 1])
else:
    print(None)
    

# 8.3. Expected Output: 6
def find_nth_largest(lst, n):
    unique_elements = sorted(set(lst), reverse=True)
    return unique_elements[n - 1] if len(unique_elements) >= n else None

print(find_nth_largest([4, 8, 6, 2, 8], 2))


# 8.4. Expected Output: 4
the_list = [4, 8, "9", 2]
nth_largest_value = 2
numeric_list = [x for x in the_list if isinstance(x, (int, float))]
if len(set(numeric_list)) >= nth_largest_value:
    sorted_list = sorted(set(numeric_list), reverse=True)
    print(sorted_list[nth_largest_value - 1])
else:
    print(None)
    

# 8.5. Expected Output: 7
import heapq

the_list = [4, 8, 2, 7, 9, 1]
nth_largest_value = 3
print(heapq.nlargest(nth_largest_value, the_list)[-1])
```

### 11. Solution(s): Print Common Elements in Two Lists

```python
# 11.1. Expected Output: [30] and [30]
# Standard Approach
common_list = []
for i in first_list:
    if i in second_list:
        if i not in common_list:
            common_list.append(i)
print(common_list)

# Optimized Solution
common_list = list(set(first_list) & set(second_list))
print(common_list)


# 11.2. Expected Output: [3]
def find_common_elements(lst1, lst2):
    return list(set(lst1) & set(lst2))

print(find_common_elements([1, 2, 3], [3, 4, 5]))


# 11.3. Expected Output: {2: (2, 1), 3: (1, 2)}
from collections import Counter

first_list = [1, 2, 2, 3]
second_list = [2, 3, 3, 4]

common_elements = set(first_list) & set(second_list)
counts = {elem: (Counter(first_list)[elem], Counter(second_list)[elem]) for elem in common_elements}
print(counts)


# 11.4. Expected Output: ['a', 'b']
first_list = ['a', 'b', 'A']
second_list = ['A', 'B', 'b']

common_list = list(set(elem.lower() for elem in first_list) & set(elem.lower() for elem in second_list))
print(common_list)


# 11.5. Expected Output: [3, 4]
first_list = [1, 2, 3, 4]
second_list = [3, 4, 5, 6]

common_list = [i for i in first_list if i in second_list and first_list.count(i) == 1]
print(common_list)
```

### 12. Solution(s): Make a Frequency Dictionary from the Elements of a List

```python
# 12.1. Expected Output: {1: 1, 2: 2, 3: 3, 4: 4} and  Counter({4: 4, 3: 3, 2: 2, 1: 1})
the_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# Standard Approach
the_dict = {}
for elem in the_list:
    if elem not in the_dict:
        the_dict[elem] = 1
    else:
        the_dict[elem] += 1
print(the_dict)


# Optimized Solution
from collections import Counter
the_dict = Counter(the_list)
print(the_dict) # Note: the Counter object displays elements in descending order of frequency by default when printed. 
      

# 12.2. Expected Output: Counter({'a': 2, 'b': 1, 'c': 1})
from collections import Counter
def frequency_dict(lst):
    return Counter(lst)

print(frequency_dict(['a', 'b', 'a', 'c']))

    
# 12.3. Expected Output: Counter({'a': 4})
from collections import Counter
the_list = ['A', 'a', 'A', 'a']
the_dict = Counter(elem.lower() for elem in the_list)
print(the_dict)


# 12.4. Expected Output: {1: 1, 2: 1, 3: 1}
the_list = [1, None, 2, None, 3, None]
the_dict = {}
for elem in the_list:
    if elem is not None:
        if elem not in the_dict:
            the_dict[elem] = 1
        else:
            the_dict[elem] += 1
print(the_dict)


# 12.5. Expected Output: Counter({'this': 2, 'is': 2, 'a': 2, 'test': 2, 'only': 1})
from collections import Counter
sentence = "this is a test this is only a test".split()
word_count = Counter(sentence)
print(word_count)
```

### 14. Solution(s): Count Elements Greater Than

```python
# 14.1. Expected Output: 3 and 3
the_list = [10, 20, 30, 40]

# Standard Approach
the_count = 0
for elem in the_list:
    if elem > 15:
        the_count += 1
print(the_count)


# Optimized Solution
the_count = sum(1 for elem in the_list if elem > 15)
print(the_count)


# 14.2. Expected Output: 2
def count_greater(lst, value):
    return sum(1 for elem in lst if elem > value)

print(count_greater([1, 2, 3, 4], 2))

        
# 14.3. Expected Output: 3
the_list = [1, 4, 5, 6, 10, 12]
the_count = sum(1 for elem in the_list if 3 < elem < 10)
print(the_count)


# 14.4. Expected Output: 2
the_list = [1, "10", 5, 15]
the_count = sum(1 for elem in the_list if isinstance(elem, (int, float)) and elem > 3)
print(the_count)

        
# 14.5. Expected Output: 2
the_list = [1, 2, 3, 4, 5]
greater_list = [elem for elem in the_list if elem > 3]
print(len(greater_list))
```

### 15. Solution(s): Difference Between Two Lists

```python
# 15.1. Expected Output: [1, 2, 3]
the_first_list = [1, 2, 3, 4, 5]
the_list_2compare = [4, 5, 6]

# Standard Approach
the_difference_list = []
for elem in the_first_list:
    if elem not in the_list_2compare:
        the_difference_list.append(elem)
print(the_difference_list)

# Optimized Solution
the_difference_list = [elem for elem in the_first_list if elem not in the_list_2compare]
print(the_difference_list)


# 15.2. Expected Output: [1, 2]
def list_difference(lst1, lst2):
    return [elem for elem in lst1 if elem not in lst2]

print(list_difference([1, 2, 3, 4], [3, 4, 5]))


# 15.3. Expected Output: [1]
the_first_list = [1, 2, 2, 3]
the_list_2compare = [2, 3]
the_difference_list = [elem for elem in the_first_list if elem not in the_list_2compare]
print(the_difference_list)


# 15.4. Expected Output: [[1, 2]]
the_first_list = [[1, 2], 3]
the_list_2compare = [3, [1, 2]]
the_difference_list = [elem for elem in the_first_list if elem not in the_list_2compare]
print(the_difference_list)

        
# 15.5. Expected Output: [1, 2]
the_first_list = [1, 2, 3, 4]
the_list_2compare = [3, 4, 5]

the_difference_list = list(set(the_first_list) - set(the_list_2compare))
print(the_difference_list)
```

### 16. Solution(s): Calculate Distance Between Two Points in 3D Space

```python
# 16.1. Expected Output: 1.7320508075688772
import math

pointA = [0, 0, 0]
pointB = [1, 1, 1]

# Calculate the sum of squared differences
addition = ((pointA[0] - pointB[0]) ** 2
            + (pointA[1] - pointB[1]) ** 2
            + (pointA[2] - pointB[2]) ** 2)

# Calculate the distance
distance = math.sqrt(addition)

# Print the result
print(distance)


# 16.2. Expected Output: 5.196152422706632
import math

def calculate_distance_3d(point1, point2):
    return math.sqrt(
        (point1[0] - point2[0]) ** 2 +
        (point1[1] - point2[1]) ** 2 +
        (point1[2] - point2[2]) ** 2
    )

print(calculate_distance_3d([1, 2, 3], [4, 5, 6]))
      

# 16.3. Expected Output: 5.196152422706632
def calculate_distance_nd(point1, point2):
    return math.sqrt(sum((a - b) ** 2 for a, b in zip(point1, point2)))

print(calculate_distance_nd([1, 2, 3], [4, 5, 6]))


# 16.4. Expected Output: 5.196152422706632
pointA = (1, 2, 3)
pointB = (4, 5, 6)
print(calculate_distance_3d(pointA, pointB))

        
# 16.5. Expected Output: 5.196152422706632
import numpy as np

pointA = np.array([1, 2, 3])
pointB = np.array([4, 5, 6])
print(np.linalg.norm(pointA - pointB))
```
