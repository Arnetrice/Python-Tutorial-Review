```python
# 2.1. The loop condition the_divisor < the_num is false for 2, so the loop does not run.
    # Since the flag variable is_prime is initialized to True, the program correctly identifies 2 as prime.


# 2.2. Add a condition at the start
# Input: User provides a number
the_num = int(input("Enter a number: "))

# Handle edge cases for 1 and negative numbers
if the_num <= 1:
    print("Not Prime")
else:
    the_divisor = 2 # Start with the smallest prime divisor

    # Check for divisors
    while the_divisor < the_num:  
        if the_num % the_divisor == 0:
            print("Not Prime")
            break
        the_divisor += 1
    else:
        print("Prime")



# 2.3. Implement square root and check
import math  # Import math to use the square root function

# Input: User provides a number
the_num = int(input("Enter a number: "))

# Handle edge cases for 1 and negative numbers
if the_num <= 1:
    print("Not Prime")
else:
    the_divisor = 2  # Start with the smallest prime divisor

    # Optimized loop: Check divisors up to the square root of the number
    while the_divisor <= math.isqrt(the_num):
        if the_num % the_divisor == 0:  # Check if divisible
            print("Not Prime")
            break
        the_divisor += 1
    else:
        print("Prime")



# 2.4. Replace the while loop with a for loop
import math  # Import math to calculate the integer square root

# Input: User provides a number
the_num = int(input("Enter a number: "))

# Handle edge cases for 1 and negative numbers
if the_num <= 1:
    print("Not Prime")
else:
    # Use a for loop to check divisors up to the square root of the number
    for the_divisor in range(2, math.isqrt(the_num) + 1):
        if the_num % the_divisor == 0:  # Check for divisibility
            print("Not Prime")
            break  # Exit the loop if a divisor is found
    else:
        # Executes only if the loop completes without hitting 'break'
        print("Prime")



# 2.5. Add a nested loop to check each number
import math

# Input: User provides an upper limit for the range
n = int(input("Enter the upper limit (n): "))

# Edge case: Handle if n is less than 2
if n < 2:
    print("There are no prime numbers in this range.")
else:
    print("Prime numbers from 2 to", n, "are:")

    # Iterate through all numbers in the range 2 to n
    for num in range(2, n + 1):  # Start at 2, end at n
        is_prime = True  # Assume the number is prime initially

        # Check divisors up to the square root of the number
        for divisor in range(2, math.isqrt(num) + 1):
            if num % divisor == 0:  # Not a prime number
                is_prime = False
                break

        # Print the number if it is prime
        if is_prime:
            print(num)
```

### 3. Solution(s): Generate the First 100 Even Numbers Using a While Loop

```python
# 3.1. Generate the First n Even Numbers
n = int(input("Enter how many even numbers to generate: "))
the_counter = 2
count = 0
while count < n:
    print(the_counter, end=" ")
    the_counter += 2
    count += 1

# 3.2.Generate Even Numbers Between Two Give Numbers such as 50 and 150
the_counter = 50
while the_counter <= 150:
    print(the_counter, end=" ")
    the_counter += 2

# 3.3. Generate Odd Numbers
the_counter = 1
count = 0
while count < 100:
    print(the_counter, end=" ")
    the_counter += 2
    count += 1

# 3.4. Generate the First n Even Numbers Using a For Loop
# Input: User provides how many even numbers to generate
n = int(input("Enter how many even numbers to generate: "))

# Use a for loop to generate the first n even numbers
for the_counter in range(2, 2 * n + 1, 2):  # Start at 2, step by 2, stop at 2*n
    print(the_counter, end=" ")

    
# 3.5. No Even Numbers
n = int(input("Enter how many even numbers to generate: "))
if n <= 0:
    print("No even numbers to display.")
else:
    the_counter = 2
    count = 0
    while count < n:
        print(the_counter, end=" ")
        the_counter += 2
        count += 1
```

### 4. Solution(s): Print the Alphabet Using a While Loop

```python
# 4.1. Print Lowercase Alphabet
start = ord('a')
end = ord('z') + 1
ascii_value = start
while ascii_value < end:
    print(chr(ascii_value), end=' ')
    ascii_value += 1
print()


# 4.2. Print a Range of Letters
start_letter = input("Enter the start letter: ").upper()
end_letter = input("Enter the end letter: ").upper()

start = ord(start_letter)
end = ord(end_letter) + 1

ascii_value = start
while ascii_value < end:
    print(chr(ascii_value), end=' ')
    ascii_value += 1
print()


# 4.3. Reverse Alphabet Order
start = ord('Z')
end = ord('A') - 1
ascii_value = start
while ascii_value > end:
    print(chr(ascii_value), end=' ')
    ascii_value -= 1
print()


# 4.4. Odd or Even Letters Only 
start = ord('A')
end = ord('Z') + 1
ascii_value = start
while ascii_value < end:
    if (ascii_value - start) % 2 == 0:  # Odd-positioned letters
        print(chr(ascii_value), end=' ')
    ascii_value += 1
print()


start = ord('A')  # ASCII value of 'A'
end = ord('Z') + 1  # ASCII value of 'Z' + 1 to include 'Z'

ascii_value = start
while ascii_value < end:
    if (ascii_value - start) % 2 != 0:  # Even-positioned letters
        print(chr(ascii_value), end=' ')
    ascii_value += 1
print()

# 4.5. Print Letters at Positions Divisible by 3
start = ord('A')  # ASCII value of 'A'
end = ord('Z') + 1  # ASCII value of 'Z' + 1 to include 'Z'

ascii_value = start
while ascii_value < end:
    if (ascii_value - start) % 3 == 0:  # Check divisibility by 3
        print(chr(ascii_value), end=' ')
    ascii_value += 1
print()
```

### 5.  Solution(s): Reverse a String Using a While Loop

```python
# 5.1. Expected Ouptut: !dlroW olleH
sentence = "Hello World!"
reversed_sentence = ''.join(sentence[i] for i in range(len(sentence) - 1, -1, -1))
print(reversed_sentence)

# 5.2. Expected Ouptut: olleH !dlroW
sentence = "Hello World!"
reversed_words = ' '.join(word[::-1] for word in sentence.split())
print(reversed_words)

# 5.3. Expected Ouptut: OLLEh
the_string = "Hello"
reversed_case_swapped = ''.join(
    the_string[i].swapcase() for i in range(len(the_string) - 1, -1, -1)
)
print(reversed_case_swapped)


# 5.4. Expected Ouptut: holle
the_string = "hello"
vowels = "aeiouAEIOU"
string_list = list(the_string)
i, j = 0, len(the_string) - 1
while i < j:
    if string_list[i] not in vowels:
        i += 1
    elif string_list[j] not in vowels:
        j -= 1
    else:
        string_list[i], string_list[j] = string_list[j], string_list[i]
        i += 1
        j -= 1
print(''.join(string_list))

# 5.5. Expected Ouptut: nohtyP
the_string = "Python"
stack = list(the_string)  # Convert string to a list (stack-like structure)
reversed_string = ""
while stack:
    reversed_string += stack.pop()  # Pop characters from the end
print(reversed_string)
```

### 6. Solution(s): Reverse Digits of a Number Using a While Loop

```python
# 6.1. Expected Ouptut: -321
the_num = int(input("Enter a number: "))
is_negative = the_num < 0  # Check if the number is negative
the_num = abs(the_num)  # Work with the absolute value
reversed_num = 0

while the_num:
    reversed_num = reversed_num * 10 + the_num % 10
    the_num //= 10\

if is_negative:
    reversed_num = -reversed_num  # Add the negative sign back

print(reversed_num)

# 6.2. Expected Ouptut: Reversed Number: 87654, Number of Digits: 5
the_num = int(input("Enter a number: "))
reversed_num = 0
digit_count = 0  # Initialize the counter for digits

while the_num:
    reversed_num = reversed_num * 10 + the_num % 10  # Reverse the number
    the_num //= 10  # Remove the last digit
    digit_count += 1  # Increment the digit count

print(f"Reversed Number: {reversed_num}, Number of Digits: {digit_count}")


# 6.3. Expected Ouptut: 54.321
the_num = input("Enter a decimal number: ")
reversed_num = ''.join(reversed(the_num))
print(reversed_num)

  
# 6.4. Expected Ouptut: Reversed Number: 54321, Sum of Digits: 15
the_num = int(input("Enter a number: "))
reversed_num = 0
digit_sum = 0

while the_num:
    digit = the_num % 10
    reversed_num = reversed_num * 10 + digit
    digit_sum += digit  # Add each digit to the sum
    the_num //= 10

print(f"Reversed Number: {reversed_num}, Sum of Digits: {digit_sum}")

# 6.5. Expected Ouptut: 6789 for input 9876
def reverse_num(num, reversed_num=0):
    if num == 0:
        return reversed_num
    return reverse_num(num // 10, reversed_num * 10 + num % 10)

the_num = int(input("Enter a number: "))
print(reverse_num(the_num))
```

### 10. Solution(s): Infinite Loop with Break Condition

```python
# 10.1. Expected Output: You entered: hello (Input #1), You entered: python (Input #2), 
    # and Exiting the loop after 2 inputs.
counter = 0
while (user_input := input('Type "exit" to quit: ').lower()) != 'exit':
    counter += 1
    print(f'You entered: {user_input} (Input #{counter})')
print(f'Exiting the loop after {counter} inputs.')


# 10.2. Expected Output: You entered: hello, You entered: python, and Exiting the loop using the keyword "quit".
    # for inputs hello, python, and quit
exit_keyword = input('Enter the custom exit keyword: ').lower()
while (user_input := input(f'Type "{exit_keyword}" to quit: ').lower()) != exit_keyword:
    print(f'You entered: {user_input}')
print(f'Exiting the loop using the keyword "{exit_keyword}".')


# 10.3. Expected Output: Input too long. Try again., You entered: hello, and Exiting the loop.
    # for inputs python12345, hello, and exit
while (user_input := input('Type "exit" to quit (max 10 characters): ').lower()) != 'exit':
    if len(user_input) > 10:
        print('Input too long. Try again.')
    else:
        print(f'You entered: {user_input}')
print('Exiting the loop.')


# 10.4. Expected Output: Empty input is not allowed. Try again., You entered: python, and Exiting the loop.
    # for inputs "", python, and exit
while (user_input := input('Type "exit" to quit: ').strip().lower()) != 'exit':
    if user_input == '':
        print('Empty input is not allowed. Try again.')
    else:
        print(f'You entered: {user_input}')
print('Exiting the loop.')

   
# 10.5. Expected Output: Countdown: 10..... Countdown: 1 Countdown complete. Exiting the loop.
countdown = 10
while countdown > 0:
    print(f'Countdown: {countdown}')
    countdown -= 1
print('Countdown complete. Exiting the loop.')
```

```python
# 11.1. Expected Output: 
the_num = int(input('Enter a number: '))
the_multiplier = 1

while the_multiplier <= 10:
    print(f'{the_num} x {the_multiplier} = {the_num * the_multiplier}')
    the_multiplier += 1


# 11.2. Expected Output: 
the_num = int(input('Enter a number: '))
start = int(input('Enter the starting multiplier: '))
end = int(input('Enter the ending multiplier: '))

while start <= end:
    print(f'{the_num} x {start} = {the_num * start}')
    start += 1

    
# 11.3. Expected Output: 
the_num = 1
while the_num <= 5:
    print(f'Multiplication Table for {the_num}')
    the_multiplier = 1
    while the_multiplier <= 10:
        print(f'{the_num} x {the_multiplier} = {the_num * the_multiplier}')
        the_multiplier += 1
    print()
    the_num += 1

    
# 11.4. Expected Output: 
the_num = 5
the_multiplier = 10

while the_multiplier >= 1:
    print(f'{the_num} x {the_multiplier} = {the_num * the_multiplier}')
    the_multiplier -= 1


# 11.5. Expected Output: 
the_num = 5
the_multiplier = 1
table = []

while the_multiplier <= 10:
    result = f'{the_num} x {the_multiplier} = {the_num * the_multiplier}'
    table.append(result)
    the_multiplier += 1

for line in table:
    print(line)
```
